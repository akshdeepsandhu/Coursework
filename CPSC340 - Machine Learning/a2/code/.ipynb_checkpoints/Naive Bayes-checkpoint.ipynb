## Fit function:


```python
 dataset = load_dataset("newsgroups.pkl")

X = dataset["X"]
y = dataset["y"]
X_valid = dataset["Xvalidate"]
y_valid = dataset["yvalidate"]
```

```python
N, D = X.shape


# Compute the number of class labels
C = 4

# Compute the probability of each class i.e p(y==c)
counts = np.bincount(y)
p_y = counts / N

# Compute the conditional probabilities i.e.
# p(x(i,j)=1 | y(i)==c) as p_xy
# p(x(i,j)=0 | y(i)==c) as p_xy
p_xy = np.zeros((D, C))

#loop over every row 
#find out class from y-vector
# loop over every column, find out binary value, increase count in p_xy
for i in range(0,N):
        c = y[i]
        for j in range(0,D):
            #loop over every column, i.e. class, if j-th value == 1, increase count
            if X[i][j] == 1:
                p_xy[j][c] += 1

#normalize by class frequency
p_xy = p_xy/counts

#set as class vars
p_y = p_y
p_xy = p_xy
```
