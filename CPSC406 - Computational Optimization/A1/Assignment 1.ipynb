## Question 2 
Consider $ f(x_1,x_2) = (4x_1^2 − x_2)^2$

i.  Plot  a  surface  plot  for  this  2-D  function  in  the  range $x_1 \in (−1,1)$  and $x_2 \in (−5,3).$ 
Is  this  function convex?  Concave?  neither?

```julia
using Plots
using LinearAlgebra

x = range(-1,stop=1,length=20);
y = range(-5,stop=3, length=20);
f1 = Float64[(4*xi^2 - yi)^2 for xi = x, yi = y ];
surface(x,y,f1,alpha=0.5);

#png(surface1, "surface1");
```

ii. The function $ f(x_1,x_2) = 2x_2^2 - 6x_2^2 + 3x_1x_2$ has a stationary point at (0,0). The Hessian at this point is: $\nabla f(0,0) = 
\left[ \begin{array}{cccc} 
0 & 0 \\
0 & -12
\end{array} \right]$. The Hessian is negative semi-definite, therefore it is unclear if this stationary point is a local maximum or a saddle point. A cotnour plot of the neighbourhood can help clear things up. 

```julia
#using Plots; pyplot()
cd("$(homedir())/Desktop")

default(size=(500,300))
x1 = x = range(-10,stop=10,length=200);
x2 = range(-10,stop=10, length=200);
f2 = Float64[2*xi2^3 - 6*xi2^2 + 3*(xi1^2)*xi2 for xi1 = x1, xi2 = x2 ];
contour_plot1 = plot(x1,x2,f2,nlevels=50,w=1)
#png(contour_plot1, "contour_plot1")
```

iii. For each stationary point, plot a surface plots for this 2-D function that zooms in on the stationary point. Visually, are they local minima, local maxima, or saddle points


```julia

x2 = range(-0.5,stop=0.5,length=30);
x3 = range(-0.5,stop=0.5, length=30);
f3 = Float64[2*xi2^3 - 6*xi2^2 + 3*(xi1^2)*xi2 for xi1 = x2, xi2 = x3 ];
surface2 = surface(x2,x3,f3, alpha=0.6)
surface_plot_0_0 = plot!(surface2[1], camera=(30, 20))
#surface_plot_0_0_side =  plot!(surface2[1], camera=(0, 0))
png(surface_plot_0_0, "surface_plot_0_0")
```
